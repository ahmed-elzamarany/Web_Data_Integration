```python
import numpy as np
# import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; 
sns.set(rc={'figure.figsize':(7,7)})
```

# Helper functions

```python
def rename(df,ISBN,Title,Authors,Rating,Pages ,Description ,Year_of_publication ,Publisher ,Genres ,Price):
    df.rename(columns = {ISBN:'ISBN'}, inplace = True) if ISBN else next
    df.rename(columns = {Title:'Title'}, inplace = True) if Title  else next
    df.rename(columns = {Authors:'Authors'}, inplace = True) if Authors  else next
    df.rename(columns = {Rating:'Rating'}, inplace = True) if Rating  else next
    df.rename(columns = {Pages:'Pages'}, inplace = True) if Pages  else next
    df.rename(columns = {Description:'Description'}, inplace = True) if Description  else next
    df.rename(columns = {Year_of_publication:'Year_of_publication'}, inplace = True) if Year_of_publication  else next
    df.rename(columns = {Publisher:'Publisher'}, inplace = True) if Publisher  else next
    df.rename(columns = {Genres:'Genres'}, inplace = True) if Genres  else next
    df.rename(columns = {Price:'Price'}, inplace = True) if Price  else next
```

# Kindle data set

```python
Kindle_df = pd.read_csv('../Datasets/Kindle_Book_Dataset.csv')
Kindle_df.info()
```

```python
rename(Kindle_df,None,'title','author','stars','pages' ,'description' ,None ,'publisher' ,None ,'price')
Kindle_df = Kindle_df[['Title','Authors','Rating','Pages' ,'Description' ,'Publisher'  ,'Price']]
Kindle_df
```

```python
null_count=(Kindle_df.isnull().sum()/len(Kindle_df))*100
df_null_count = pd.DataFrame({'columns':Kindle_df.columns, 'Pertentage of nulls (%)':null_count.values})
sns.barplot(x = 'Pertentage of nulls (%)',
            y = 'columns',
            data = df_null_count)
```

# Goodreads and Book Recommendation Dataset

```python
Goodreads_Recommendation_df = pd.merge(Recommendation_df, Goodreads_df,  left_on=['Title','Authors'], right_on = ['Title','Authors']) 
Goodreads_Recommendation_df.drop_duplicates(subset= 'ISBN',keep = 'first', inplace = True)
Goodreads_Recommendation_df
```

We have 32831 common entities between Goodreads and Book Recommendation Dataset

```python
All_df = pd.merge(Kindle_df, Goodreads_Recommendation_df, how ='outer',left_on=['Title','Authors'], right_on = ['Title','Authors']) 
All_df
```

```python
All_df["Rating_x"].update(All_df.pop("Rating_y"))
All_df["Pages_x"].update(All_df.pop("Pages_y"))
All_df["Description_x"].update(All_df.pop("Description_y"))
All_df["Publisher_x"].update(All_df.pop("Publisher_y"))
All_df
```

```python
null_count=(All_df.isnull().sum()/len(All_df))*100
df_null_count = pd.DataFrame({'columns':All_df.columns, 'Pertentage of nulls (%)':null_count.values})
sns.barplot(x = 'Pertentage of nulls (%)',
            y = 'columns',
            data = df_null_count)
null_count
```
