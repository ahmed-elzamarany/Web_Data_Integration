```python
import xml.etree.ElementTree as ET
import numpy as np
import pandas as pd
```

```python
goodreads_kindle_correspondences = pd.read_csv("./correspondences/goodreads_kindle_correspondences.csv")
goodreads_kindle_correspondences
```

```python
recommendation_goodreads_correspondences = pd.read_csv("./correspondences/recommendation_goodreads_correspondences.csv")
recommendation_goodreads_correspondences
```

```python
goodreads_df = pd.merge(goodreads_kindle_correspondences, recommendation_goodreads_correspondences,  left_on=['id_goodreads'], right_on = ['id_goodreads']) 
goodreads_df
```

```python
goodreads_df = goodreads_df.drop(goodreads_df[goodreads_df['value_x'] != 1.0].index)
goodreads_df = goodreads_df.drop(goodreads_df[goodreads_df['value_y'] != 1.0].index)
goodreads_df
```

```python
recommendation_kindle_correspondences = pd.read_csv("./correspondences/recommendation_kindle_correspondences.csv")
recommendation_kindle_correspondences
```

```python
all_df=pd.merge(goodreads_df, recommendation_kindle_correspondences,  left_on=['id_recommendation','id_kindle'], right_on = ['id_recommendation','id_kindle'])
all_df = all_df.drop(all_df[all_df['value'] != 1.0].index)
all_df
```

```python
droped=all_df.drop_duplicates(subset=['id_goodreads'],keep='first')
droped=droped.drop_duplicates(subset=['id_kindle'],keep='first')
droped=droped.drop_duplicates(subset=['id_recommendation'],keep='first')
droped.reset_index( inplace=True)
droped.drop("index",1,inplace=True)
droped
```

```python
def extractXML(name):
    s=name.split("_")
    tree = ET.parse(s[0]+".xml")    
    root = tree.getroot()
    for child in root:
        if child.findall('id')[0].text==name:
            return child
        
```

```python
root = ET.Element("books")
for i in droped.index:
    e1=extractXML(droped.loc[i]["id_goodreads"])
    root.insert(0,e1)
    e2=extractXML(droped.loc[i]["id_kindle"])
    root.insert(0,e2)
    e3=extractXML(droped.loc[i]["id_recommendation"])
    root.insert(0,e3)
```

```python
print(ET.tostring(root, encoding='utf8').decode('utf8'))
```

```python
tree=ET.ElementTree(root)
with open('gold_H.xml', 'wb') as f:
    tree.write(f)
```
