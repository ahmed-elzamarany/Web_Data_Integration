```python
import xml.etree.ElementTree as ET

def elements_equal(e1, e2):
    if type(e1) != type(e2):
        return False
    if e1.tag != e2.tag: return False
    if e1.text != e2.text: return False
    if e1.tail != e2.tail: return False
    if e1.attrib != e2.attrib: return False
    if len(e1) != len(e2): return False
    return all([elements_equal(c1, c2) for c1, c2 in zip(e1, e2)])

def id_xml(tree,name):
    root = tree.getroot()
    s=name+"_"
    i=1
    for child in root:
        c = ET.Element("id")
        c.text = s+str(i)
        child.insert(0, c)
        i+=1
    return ET.ElementTree(root)

def remove_duplicates(tree):
    root = tree.getroot()
    prev = None
    elems_to_remove = []
    for page in root:                     # iterate over pages
        if elements_equal(page, prev):
            print("found duplicate: %s" % ET.tostring(page, encoding='utf8').decode('utf8'))   # equal function works well
            elems_to_remove.append(page)
        prev = page
    for elem_to_remove in elems_to_remove:
        root.remove(elem_to_remove)
    return ET.ElementTree(root)

def remove_nulls_author(tree):
    root = tree.getroot()
    prev = None
    elems_to_remove = []
    for page in root:                     # iterate over pages
        if page.findall('authors')[0].findall("author")[0].findall("name")==[]:
            print("found null: %s" % ET.tostring(page, encoding='utf8').decode('utf8'))   # equal function works well
            elems_to_remove.append(page)
    for elem_to_remove in elems_to_remove:
        root.remove(elem_to_remove)
    return ET.ElementTree(root)
```

```python
def Pipline(name):
    tree = ET.parse("../Data_Translation/output/"+name+".xml")
    tree = remove_duplicates(tree)
    tree = remove_nulls_author(tree)
    tree = id_xml(tree,name)
    with open(f'./data-cleaned/{name}.xml', 'wb') as f:
        tree.write(f)
```

```python
Pipline("kindle")
Pipline("recommendation")
Pipline("goodreads")
```
